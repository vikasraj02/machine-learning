```python
file_list = []
for i in range(65,91):
    l = glob.glob(file_path+"\\"+ chr(i))
    file_list.append(l)
```

```python
file_inner_img = []
for i in file_list:
    file_inner_img.append(glob.glob(i[0]+"\\*.png"))
```

```python
file_inner_img
```

```python
img_array = []
lable = []
for i in file_inner_img:
    for j in tqdm(i):
        #print(j)
        img =image.imread(j)
        img_flatten = img.flatten()
        value = np.asarray(img_flatten)
        img_array.append(value) 
        re_lable = re.findall(r"(\w)-\d+",j)
        lable.append(re_lable[0])
        
        
```

```python
df = pd.DataFrame(img_array)
```

```python
df["lable"] = lable
```

```python
df
```

```python
df.to_csv("image_df")
```

```python
df = pd.read_csv("image_df")
```

```python
df.head()
```

```python
df.drop("Unnamed: 0",axis = 1 , inplace = True)
```

# test, train split

```python
x = df.select_dtypes(include = ["float"])
```

```python
y = df.select_dtypes(include = "object")\

y.head()
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train = train_test_split(x,y,train_size=0.75, random_state = 50)
```

# Algo

# logisticRegression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

```python
y_test_pred = classifier.predict(x_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_test_pred)
```

# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
```

```python
y_test_pred_random_forest = classifier.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_pred_random_forest)
```

# decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train)
```

```python
y_test_pred_decision = classifier.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_pred_decision)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(x_train,y_train)
```

```python
y_test_pred_knn = classifier.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_pred_knn)
```

# svc

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(x_train,y_train)
```

```python
y_test_pred_svc = classifier.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_test_pred_svc)
```
